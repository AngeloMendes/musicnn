```python
from musically_motivated_CNN import musiCNN
taggram, tags = musiCNN.predict(file_name='./audio/TRWJAZW128F42760DD_test.mp3', model='MTT')
```

Out of `musiCNN.predict()` you get a **taggram** and its associated **tags**:

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 150
import numpy as np

fig, ax = plt.subplots()
pos = np.arange(len(tags))
ax.title.set_text('Taggram')
ax.title.set_fontsize(8)
ax.imshow(taggram.T, interpolation=None, aspect="auto")
ax.set_yticks(pos)
ax.set_yticklabels(tags, fontsize=6)
ax.tick_params(axis="x", labelsize=6)
plt.show()
```

And from the taggram, we can compute the most likely tags of this by simply averaging through time:

```python
tags_likelihood_mean = np.mean(taggram, axis=0)

fig, ax = plt.subplots()
pos = np.arange(len(tags))
ax.title.set_text('Predicted tags (mean of taggram)')
ax.bar(pos, tags_likelihood_mean)
ax.set_xticks(pos)
ax.set_xticklabels(tags, rotation=90)
plt.show()
```

```python


# From the taggram one can estimate the most popular tags
topN = 3
print(file_name)
for tag_index in tags_likelihood_mean.argsort()[-topN:][::-1]:
    print(' - [mean] ' + tags[tag_index])
```
